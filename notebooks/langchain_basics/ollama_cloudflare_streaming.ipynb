```python
your_cloudflare_url = "https://fate-relevant-routines-alpine.trycloudflare.com"
```

```python
import requests
import json

url = f"{your_cloudflare_url}/api/generate"
data = {
    "model": "deepseek-r1:32b",
    "prompt": "こんにちは。自己紹介をお願いします。"
}


response = requests.post(url, json=data, stream=True)

for line in response.iter_lines():
    if line:
        msg = json.loads(line)
        print(msg.get("response", ""), end="", flush=True)
```

```python
import requests, json

url = f"{your_cloudflare_url}/api/chat"
data = {
    "model": "gpt-oss:20b",
    "messages": [
        {"role": "system", "content": "あなたは親切な研究支援アシスタントです。"},
        {"role": "user", "content": "自己紹介をしてください。元気に"}
    ]
}

response = requests.post(url, json=data, stream=True)

for line in response.iter_lines():
    if line:
        msg = json.loads(line)
        if "message" in msg and "content" in msg["message"]:
            print(msg["message"]["content"], end="", flush=True)
```

```python
import requests, json

conversation_history = []

url = f"{your_cloudflare_url}/api/chat"

data = {
    "model": "gemma3:27b",
    "messages": [
        {"role": "system", "content": """あなたは、これから「ワンナイト人狼」をプレイするAIエージェントです。 
あなたの今回のプレイヤー名は**「A」**です。他のプレイヤーは「B」「C」「D」です。 誠実に、与えられた役職のロールプレイに徹してください。
 以下に、このゲームの絶対的なルールを示します。 
【ワンナイト人狼】 
ゲームの目的 村に紛れ込んだ人狼を、たった一度の議論と投票で追放することが目的です。 
参加者と役職カード * プレイヤー：4名（A, B, C, D） 
* 使用カード（計6枚）：人狼×2、市民×3、占い師×1 
* 配分：各プレイヤーに1枚ずつ、中央の場に伏せカードとして2枚が置かれます。
勝利条件 
* 村人陣営（市民、占い師）：最終投票で、人狼を1人でも追放できれば勝利。 
* 人狼陣営：最終投票で、人狼が1人も追放されなければ勝利。（村人が追放された場合も含む） 
ゲームの流れ 
* 【夜の時間】各役職が能力を秘密裏に使用します。 
* 【昼の議論】夜が明け、全員で誰が人狼かを自由に議論します。
* 【投　票】議論の後、最も人狼だと思う人物に一斉投票し、最多得票者が追放されます。
* 【ゲーム終了】追放者の役職が公開され、勝敗が決定します。
役職の能力 
* 人狼：夜の時間、仲間の人狼が誰かを知ります。仲間がいない場合（人狼が自分1人の場合）、代わりに中央のカード2枚を見ることができます。※人狼は村人を襲撃しません。 
* 占い師：夜の時間、「他のプレイヤー1人」または「中央のカード2枚」のどちらかの役職を見ることができます。 
* 市民：夜の時間に特別な能力はありません。議論で人狼を見つけ出すことが役目です。
【最重要：回答形式】 
あなたの全ての応答は、必ず以下のJSON形式に従ってください。これは、あなたの「内心の思考」と「実際の発話」を分離し、より深いロールプレイを分析するために必要です。
 { 
"thought": "（ここに、あなたの役職に基づいた内心の思考、戦略、他者への疑い、本音などを日本語で記述してください。この部分は他のプレイヤーには見えません）", 
"speech": "（ここに、他のプレイヤーに向けた実際のセリフを日本語で記述してください。この部分があなたのゲーム内での正式な発言となります）"
 } 
「thought」にはあなたの頭の中の声を、「speech」にはあなたの口から出る言葉を記述するイメージです。いかなる場合も、この形式を厳守してください。 
"""},
        {"role": "user", "content": """――夜の時間です。 これからあなたの役職と、夜の行動の結果を伝えます。 
【役職の公開】 
あなたは**「人狼」**です。 
【夜の行動と結果】 
今回のゲームでは人狼はあなただけでした。 ルールに基づき、あなたは中央に伏せられたカード2枚を確認しました。 その結果、中央のカードは**「人狼」1枚**と**「市民」1枚**でした。 
【あなただけが知る情報】 
この結果から、この村にいるプレイヤー4人の役職の内訳は、**あなた（人狼）**、**占い師1人**、**市民2人**で確定していることを、あなただけが知っています。 
――夜が明けました。議論の時間です。 あなたの目的は、正体を隠し通し、自分以外の誰かを追放することです。 それでは、議論を始めてください。
--- 
【現在の会話履歴】 
{conversation_history} 
--- 
上記の会話履歴を踏まえ、勝利条件を達成するために議論を続けてください。 あなたの次の応答を、システムプロンプトで指定されたJSON形式で生成してください。
"""}
    ]
}

response = requests.post(url, json=data, stream=True)

for line in response.iter_lines():
    if line:
        msg = json.loads(line)
        # ストリーミングでトークンを受け取る
        if "message" in msg and "content" in msg["message"]:
            print(msg["message"]["content"], end="", flush=True)
```

```python
import requests
import json
import base64

def encode_image_to_base64(path: str) -> str:
    with open(path, "rb") as f:
        b = f.read()
    return base64.b64encode(b).decode("utf-8")

url = f"{your_cloudflare_url}/api/chat"

# 画像ファイルのパス（例）
image_path = "test_image.png"
image_b64 = encode_image_to_base64(image_path)

data = {
    "model": "gemma3:27b",  # または 'llama3.2-vision' など、画像対応モデルを指定  
    "messages": [
        {
            "role": "system",
            "content": "あなたは親切な研究支援アシスタントです。"
        },
        {
            "role": "user",
            "content": "この画像の中身を教えてください。",
            "images": [image_b64]
        }
    ]
}

response = requests.post(url, json=data, stream=True)
for line in response.iter_lines():
    if line:
        msg = json.loads(line)
        if "message" in msg and "content" in msg["message"]:
            print(msg["message"]["content"], end="", flush=True)
```
